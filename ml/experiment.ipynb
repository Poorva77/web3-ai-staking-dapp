# ML Mini-Experiment: Staking Risk Prediction

This notebook shows a small machine learning experiment for a Web3 staking DApp.

The model predicts a simple risk level from staking behavior.

## 1. Import Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

## 2. Create Sample Dataset

For this internship task, I created a small sample dataset. In a real project, this data can come from blockchain transactions.

```python
data = {
    "stake_amount": [10, 25, 50, 5, 100, 75, 15, 200, 8, 60, 120, 30],
    "staking_days": [30, 60, 90, 5, 180, 120, 15, 365, 7, 100, 220, 45],
    "previous_withdrawals": [1, 1, 0, 5, 0, 1, 4, 0, 6, 2, 0, 3],
    "wallet_age_days": [100, 180, 300, 20, 500, 400, 45, 900, 15, 250, 700, 120],
    "risk_level": ["Medium", "Low", "Low", "High", "Low", "Low", "High", "Low", "High", "Medium", "Low", "Medium"]
}

df = pd.DataFrame(data)
df
```

## 3. Prepare Input and Output

```python
X = df[["stake_amount", "staking_days", "previous_withdrawals", "wallet_age_days"]]
y = df["risk_level"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)
```

## 4. Train Model

I used a Decision Tree Classifier because it is simple and easy to explain.

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

## 5. Test Model

```python
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("\nClassification Report:")
print(classification_report(y_test, y_pred, zero_division=0))
```

## 6. Sample Prediction

Example: a user stakes 12 tokens for 10 days, has 4 previous withdrawals, and wallet age is 30 days.

```python
sample_user = pd.DataFrame([
    {
        "stake_amount": 12,
        "staking_days": 10,
        "previous_withdrawals": 4,
        "wallet_age_days": 30
    }
])

prediction = model.predict(sample_user)
print("Predicted Risk Level:", prediction[0])
```

## Conclusion

The model gives a simple risk category from staking behavior.

This can be improved later by using real wallet transactions, larger datasets, and better evaluation.
